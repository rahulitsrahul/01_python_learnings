```python

from matplotlib import pyplot as plt
import numpy as np
```

```python
X = np.linspace(-2*np.pi, np.pi, 30)
C = np.cos(X)
S = np.sin(X)
```

```python
# Different elements within a plot
fig = plt.figure(figsize=(8,4), dpi=80)
ax = fig.add_subplot(1,1,1)  # Considering one image in the entire space
ax.grid(True, color='k', linestyle='--', linewidth='0.1')
ax.plot(X, S, '*g-', label='Sin')    # Marker, Color, Line
ax.plot(X, C, '<y-.', label='Cos')
ax.set_xlabel("Time")
ax.set_ylabel("Span")
ax.set_title("TRIG functions")
ax.legend()    # Show the legends
overall_median=0.5  # 
ax.fill_between(X, C, overall_median, color='b', alpha=0.1)  # Fills the area under the curve 'Cos"

plt.tight_layout() # to provide padding to the text elements
plt.show()
```

```python
fig = plt.figure(figsize=(8, 4), dpi=120)

ax1 = fig.add_subplot(121)    # (no. of rows, no.of cols, selection index)
ax2 = fig.add_subplot(122)

ax1.grid('on')
ax1.plot(X, C, 'b--o')

ax2.plot(X, S, 'r')
ax2.axhline(0, color='k', linestyle='--', label='y=0')

plt.show()
```

```python
fig = plt.figure(figsize=(8, 4), dpi=120)
plt.subplot(2, 1, 1)    # (no. of rows, no.of cols, selection index)
plt.grid('on')
plt.plot(X, C, 'b--o')
plt.axvspan(-2, 1, color='lightgray')
plt.subplot(2, 1, 2)
plt.plot(X, S, 'r')
plt.axhline(0)
plt.axvline(0.5)
plt.show()
```

```python
# Gridspec
import matplotlib.gridspec as gridspec
fig = plt.figure(figsize=(10, 5), dpi=80)
gs = gridspec.GridSpec(3,3)
ax1 = plt.subplot(gs[0, :])
ax2 = plt.subplot(gs[1, 0:2])
ax3 = plt.subplot(gs[2, 1:3])
ax1.plot(X, C)
ax2.plot(X, S)
ax3.plot(X, C+S)
plt.show()
```
