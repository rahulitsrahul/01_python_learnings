```python

from matplotlib import pyplot as plt
import numpy as np
```

```python
X = np.linspace(-2*np.pi, np.pi, 30)
C = np.cos(X)
S = np.sin(X)
```

```python
# Different elements within a plot
fig = plt.figure(figsize=(8,4), dpi=80)
plt.subplot(1,1,1)  # Considering one image in the entire space
plt.grid(True, color='k', linestyle='--', linewidth='0.1')
plt.plot(X, S, '*g-', label='Sin')    # Marker, Color, Line
plt.plot(X, C, '<y-.', label='Cos')
plt.xlabel("Time")
plt.ylabel("Span")
plt.title("TRIG functions")
plt.legend()    # Show the legends
overall_median=0.5  # 
plt.fill_between(X, C, overall_median, color='b', alpha=0.1)  # Fills the area under the curve 'Cos"
plt.tight_layout() # to provide padding to the text elements
plt.show()
```

```python
fig = plt.figure(figsize=(8, 4), dpi=120)
plt.subplot(2, 1, 1)    # (no. of rows, no.of cols, selection index)
plt.grid('on')
plt.plot(X, C, 'b--o')
plt.subplot(2, 1, 2)
plt.plot(X, S, 'r')
plt.show()
```

```python
fig = plt.figure(figsize=(8, 4), dpi=120)
plt.subplot(2, 1, 1)    # (no. of rows, no.of cols, selection index)
plt.grid('on')
plt.plot(X, C, 'b--o')
plt.axvspan(-2, 1, color='lightgray')
plt.subplot(2, 1, 2)
plt.plot(X, S, 'r')
plt.axhline(0)
plt.axvline(0.5)
plt.show()
```

```python
# Gridspec
import matplotlib.gridspec as gridspec
fig = plt.figure(figsize=(10, 5), dpi=80)
gs = gridspec.GridSpec(3,3)
ax1 = plt.subplot(gs[0, :])
plt.plot(X, C)
ax2 = plt.subplot(gs[1, 0:2])
plt.plot(X, S)
ax3 = plt.subplot(gs[2, 1:3])
plt.plot(X, C+S)
plt.show()
```
