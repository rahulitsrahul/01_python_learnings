```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Create a figure and a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot
axs[0, 0].plot(x, y1, 'r', label='sin(x)')
axs[0, 0].set_title('Sine Function')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('sin(x)')
axs[0, 0].axhline(0, color='k', linestyle='--', label='y=0')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Second subplot
axs[0, 1].plot(x, y2, 'g', label='cos(x)')
axs[0, 1].set_title('Cosine Function')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('cos(x)')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Third subplot
axs[1, 0].plot(x, y3, 'b', label='tan(x)')
axs[1, 0].set_title('Tangent Function')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('tan(x)')
axs[1, 0].legend()
axs[1, 0].grid(True)
axs[1, 0].set_ylim(-10, 10)  # Limit y-axis to avoid extreme values

# Fourth subplot
dot_size = 0.5
axs[1, 1].scatter(x, y4, c='m',s=dot_size, label='exp(x)')
axs[1, 1].set_title('Exponential Function')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('exp(x)')
axs[1, 1].legend()
axs[1, 1].grid(True)
axs[1, 1].set_ylim(0, 1000)  # Limit y-axis to avoid extreme values

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
