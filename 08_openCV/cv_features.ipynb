```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img_gray = cv2.imread("bw_grad.jpg")
plt.imshow(img_gray)
```

```python
# Thresholding
fig = plt.figure(figsize=(16,8), dpi=80)
plt.subplot(2,3,1)
plt.title("Original img")
plt.imshow(img_gray)

_, th1 = cv2.threshold(img_gray, 150, 255, cv2.THRESH_BINARY)
plt.subplot(2,3,2)
plt.title("Binary")
plt.imshow(th1)

_, th2 = cv2.threshold(img_gray, 150, 255, cv2.THRESH_BINARY_INV)
plt.subplot(2,3,3)
plt.title("Binary_Inverse")
plt.imshow(th2)

_, th2 = cv2.threshold(img_gray, 150, 255, cv2.THRESH_TOZERO)
plt.subplot(2,3,4)
plt.title("Thresh to Zero")
plt.imshow(th2)
```
