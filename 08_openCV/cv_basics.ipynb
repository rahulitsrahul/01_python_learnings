Python OpenCV Basics

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('lena.png', 1)
print(img.shape)
```

```python
img_1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_1)
```

```python
img_gray = cv2.imread('lena.png', 0)
print(img_gray.shape)
plt.imshow(img_gray, cmap='gray', vmin=0, vmax=255)
```

Draw Geometric Sahpes on image

```python
img_a = img_1
img_a = cv2.line(img_a, (0,0), (255, 255), (0, 200, 0), 10) # (image, start_point, end_point, color, thickness)
img_a = cv2.arrowedLine(img_a, (0,256), (255, 400), (200, 200, 0), 10) # (image, start_point, end_point, color, thickness)
img_a = cv2.rectangle(img_a, (100, 100), (400, 400), (0, 0, 100), 4) # (image, top left corner, bottom right corner, color, thickness)
img_a = cv2.circle(img_a, (400, 200), 60, (200, 0, 0), -1) # image, center_coordinates, radius, color, thickness-> -1 means fill 
plt.imshow(img_a)
```

```python
# CONvert image to HSV plane
img_hsv = cv2.cvtColor(img_1, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
```

```python
plt.imshow(img)
print(img.shape)
b, g, r = cv2.split(img)
img_rgb = cv2.merge((r, g, b))
plt.imshow(img_rgb)
    
```

```python
# Crop image
img_crop = img_rgb[100:400, 200:500]
plt.imshow(img_crop)
```

```python
# Add two images with weights
img_add = cv2.addWeighted(img_rgb, 0.7,  img_a, 0.3, 0)
plt.imshow(img_add)
```
