```python
from PIL import Image
img_name = 'img_1.jpg'
img = Image.open(img_name)
# img.show() # Show the image outside of the notebook
display(img)

# Rotate image
img = img.rotate(angle=20, fillcolor=(0, 0, 255))
# Resize Image
img = img.resize((int(img.width/2), int(img.height/2)))
display(img)
```

```python
# Create an image
width = img.width
height = img.height
im_1 = Image.new("RGB", (width, height), (0, 255, 0))
display(im_1)
```

```python
# Blend two images
img_2 = Image.blend(img, im_1, alpha=0.5)
display(img_2)
```

```python
# Crop an image
im_3 = img.crop(box=(40, 50, 250, 150)) # x, y, width, height
display(img)
display(im_3)
```

```python
# Convert
im_4 = img.convert('L') # 'L' -> Grayscale, 'RGB' -> to rgb
display(im_4)
```
